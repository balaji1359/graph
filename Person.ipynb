```python
MalePersons = []

for _ in range(10):
    MalePersons.append(
    {    
        "FirstNname" : fake.first_name_male(), 
        "LastName": fake.last_name_male(), 
        "DOB" : fake.date_of_birth(),
        "City" : fake.city(),
        "State" : fake.state(),
        "Country" : fake.country(),
        "PhoneNumber" : fake.phone_number(),
        "Email" : fake.email()
    }
    )
```

```python
FemalePersons = []

for _ in range(10):
    FemalePersons.append(
    {    
        "FirstNname" : fake.first_name_female(), 
        "LastName": fake.last_name_female(), 
        "DOB" : fake.date_of_birth(),
        "City" : fake.city(),
        "State" : fake.state(),
        "Country" : fake.country(),
        "PhoneNumber" : fake.phone_number(),
        "Email" : fake.email()
    }
    )
```

```python
Fpersons = pd.DataFrame.from_dict(FemalePersons)
```

```python
Mpersons = pd.DataFrame.from_dict(MalePersons)
```

```python
Fpersons["Gender"] = "female"
Mpersons["Gender"] = "male"
```

```python
Mpersons
```

```python
pd.concat([Fpersons, Mpersons], ignore_index=True).to_csv("Persons.csv", index=False)
```

```python
import csv
with open('Persons.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile, escapechar="\\")
    query = ""
    for row in reader:
        s = "CREATE (:Person {"
        for k in row:
            s += f'{k}:"{row[k]}", '
        s = s[:-2] + '})\n'
        query += s 
    print(query)
```
